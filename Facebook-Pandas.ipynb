## Funciones de agregación

```python

fanpage.groupby(['type']).agg({'likes': ['min', 'max', 'mean', 'median'], 'comments': ['min', 'max', 'mean', 'median']})
```

```python
%matplotlib inline
fanpage.groupby(['type']).agg({'likes': ['min', 'max', 'mean', 'median']}).plot()
```

```python
from functools import reduce
import math

def test_sum(series):
       return reduce (lambda x, y: x+ y, series)
def prod_agg(series):
       return reduce (lambda x, y: int(x)*int(y), series)

def count_nulls(series):
    return len(series)-series.count()
def geo_mean(series):
       return math.pow(reduce (lambda x, y: int(x)*int(y), series),1/len(series))

%matplotlib inline
fanpage.groupby('type').agg({'likes': ['sum','mean',test_sum,count_nulls,geo_mean]}).plot()
#Ejercicio 3. Implementar la función de agregación producto y media geométrica
```
